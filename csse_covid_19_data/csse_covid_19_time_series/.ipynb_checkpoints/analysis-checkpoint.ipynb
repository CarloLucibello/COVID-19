```julia
using CSV, DataFrames, Plots
using Dates

## READ TIME SERIES INTO DATAFRAMES
confirmed = CSV.read("time_series_19-covid-Confirmed.csv") |> DataFrame
deaths = CSV.read("time_series_19-covid-Deaths.csv") |> DataFrame
recovered = CSV.read("time_series_19-covid-Recovered.csv") |> DataFrame

for df in [confirmed, deaths, recovered]
    rename!(df, "Country/Region"=>"country", "Province/State"=>"province")
end

## IMPORTANT DATES
date_lockdown_hubei = Date(2020, 1, 23)
date_lockdown_lombardia = Date(2020, 3, 7)
china_shift = Dates.Day(36) # CHINA-ITALY DELAY

confirmed
```

```julia
## This cell groups the data by countries (aggregate provinces)
# and then transposes the dataframes,
## so that in the end rows are dates and columns are country names

function to_date(s)
    s = String(s)
    m, d, y = parse.(Int, split(s,"/"))
    Date(2000+y, m, d) 
end

confirmed_country = by(confirmed, :country, 
    (c => sum for c in names(confirmed)[5:end])...)
confirmed_country = hcat(
    DataFrame(dates=to_date.(names(confirmed)[5:end])),
        DataFrame(collect(Matrix(confirmed_country[:,2:end])'), 
             Symbol.(confirmed_country[:,:country]|>collect)))

deaths_country = by(deaths, :country, 
    (c => sum for c in names(deaths)[5:end])...)
deaths_country = hcat(
    DataFrame(dates=to_date.(names(deaths)[5:end])),
        DataFrame(collect(Matrix(deaths_country[:,2:end])'), 
             Symbol.(deaths_country[:,:country]|>collect)))

confirmed_hubei = filter(r->r.province=="Hubei", dropmissing(confirmed))[:,5:end] |> Matrix |> vec
deaths_hubei = filter(r->r.province=="Hubei", dropmissing(deaths))[:,5:end] |> Matrix |> vec

for df in [confirmed_country, deaths_country]
    rename!(df, "Mainland China"=>"China")
end
```

```julia
### EXPONENTIAL FIT DEATHS
using LsqFit

linear(x, p) = p[1] .+ x .* p[2]
dmin = Date(2020,2,24)
dmax= Date(2020,3,7)

y = df[dmin .<= df.dates .<= dmax,:Italy]
y = log.(y)
x = Dates.value.(df[dmin .<= df.dates .<= dmax,:dates])
fit = curve_fit(linear, x, y, [1., 1.])
fdeaths(x) = exp.(linear(x, fit.param))
fit.param
```

```julia
## PLOT CONFIRMED
df = confirmed_country

p = plot(yaxis=(:log, (10^2,10^5), "COVID-19 Confirmed"), 
    legend=:bottomright, xrotation=60)
xlims!(Dates.value.((Date(2020,2,14), Date(2020,4,14))))
xticks!(Dates.value.([Date(2020,3,1), today(), Date(2020,4,1)]))

plot!(df.dates, df.Italy, labels="Italy", marker=:auto)
plot!(df.dates .+ china_shift, df.China, labels="China - $(china_shift)", marker=:auto)
savefig("plots/confirmed_italy_vs_china_$(today()).png")
p
```

```julia
## PLOT deaths
df = deaths_country

p = plot(yaxis=(:log, (10, 10^4),  "COVID-19  Deaths"), 
         legend=:bottomright, xrotation=40)
xlims!(Dates.value.((Date(2020,2,14), Date(2020,4,14))))
xticks!(Dates.value.([Date(2020,3,1),Date(2020,3,7), Date(2020,4,1)]))
scatter!(df.dates .+ china_shift, df.China, labels="China +$(china_shift)", 
    color=:red, marker=:square, malpha=0.5,)
scatter!(df.dates, df.Italy, labels="Italy", 
    color=:blue, marker=:circle)
vline!([date_lockdown_hubei + china_shift], labels="Hubei lockdown +$(china_shift)",
    color=:red, ls=:dash)
vline!([date_lockdown_lombardia], labels="Lombardia lockdown", 
    color=:blue, ls=:dash)

delta_lock = date_lockdown_lombardia - (date_lockdown_hubei + Dates.Day(36))
date_china_slow = Date(2020,3,9) |> Dates.value
date_italy_slow = Date(2020,3,9)+ delta_lock |> Dates.value
plot!(fdeaths, color="black", labels=" Italy exponential fit")
scatter!([date_china_slow], [fdeaths(date_china_slow)], labels="China slowdown", 
    marker=:star, markersize=6, color=:red)
scatter!([date_italy_slow], [fdeaths(date_italy_slow)], labels="Italy slowdown (predicted)", 
    marker=:star, markersize=6, color=:blue)

savefig("plots/deaths_italy_vs_china_$(today()).png")
p
```
